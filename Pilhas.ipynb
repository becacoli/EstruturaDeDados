````python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None

    def prepare(self, data):
        node = Node(data)
        node.next = self.head
        self.head = node

    def delete(self, data):
        temp = self.head

        while temp.next is not None:
            if temp.next.data == data:
                break
            temp = temp.next

        temp.next = temp.next.next

    def append(self, data):
        node = Node(data)

        # Lista está vazia
        if self.head is None:
            self.head = node
        else:
            temp = self.head
            while temp.next is not None:
                temp = temp.next
            temp.next = node

    def print(self):
        temp = self.head
        while temp is not None:
            print(temp.data)
            temp = temp.next

def main():
    ll = LinkedList()
    ll.append(10)
    ll.append(11)
    ll.append(12)
    ll.append(13)
    ll.append(14)
    ll.prepare(9)
    ll.delete(13)
    ll.print()
main()
````

````python
class Twitter:
  twitters = []  #lista que vai servir como feed
  top = 0

  def __init__(self,tamanho):
    self.twitters = [None] * tamanho
  
  #POSTANDO
  def postar(self, post):
    if self.top < len(self.twitters):
      self.twitters[self.top] = post
      self.top += 1
    else:
      print("não é possível inserir mais posts")

  #FIXANDO NA TELA
  #PEGAR O INDICE DO TWITTER QUE EU QUERO FIXAR E COLOCAR ELE COMO O ULTIMO INDICE P APARECER NO TOPO
  def fixar(self, indice_post):
    post = self.twitters[indice_post]               #SALVAR O INDICE DO POST EM UMA VARIAVEL
    if indice_post < self.top:
       self.twitters.insert(self.top, post)         #INSERT PARA INSERIR E DEPOIS DELETAR O INDICE -> INSERT: NO INDICE DE SELF.TOP VAMOS COLOCAR O POST
       del self.twitters[indice_post]
    else:
      print("Não dá para fixar este tweet")

  #IMPRIMINDO NA TELA 
  def imprimir(self):
    #imprimindo cada item do topo até o 0 a partir da decrementação do top 
      for post in reversed(self.twitters):
          print(post)
````

````python
def validate_expression(exp):
    stack = []

    #dicionario com a abertura de cara expressao e o fechamento 
    opening = {"(", "[", "{"}
    closing = {")", "]", "}"}

    pairs = {"(": ")", "[": "]", "{": "}"} #dicionario com os pares

    for token in exp:
        if token in opening:
          print("opening: " + token)
          stack.append(token)
        
        elif token in closing:
            print("closing: " + token)
              # se stack está vazia
            if not stack:
              return 0
          
            else:
                top = stack[-1]
                print("Top: " + top)
                if pairs[top] == token:
                  print("Pair[top]: " + token)
                  stack.pop()               #ao achar os pares, retiramos da stack
                else:
                  return 0

    # a lista não está vazia
    if stack:
      return 0
        
    return 1

expression = input("Expression: ")
print(validate_expression(expression))
````

**Implemente uma pilha para armazenar
números e escreva uma função para ordenar os
valores da pilha em ordem crescente. Você pode
utilizar uma pilha auxiliar para resolver o
problema, mas não pode utilizar outro tipo de
estrutura de dados.**

````python
class Celula:
    item = None
    next = None

    def __init__(self, item):
        self.item = item

class Stack:    #pilha
    top = None
    tam = 0

    def insert(self, item):
      celula = Celula(item)    #chamando a classe celula 
      celula.next = self.top
      self.top = celula
      self.tam += 1

    def remove(self):
        if self.tam > 0:
          celula = self.top
          item = celula.item
          self.top = celula.next
          del celula
          self.tam -= 1
          return item
        else:
          print("ERRO: NÃO TEM ELEMENTOS NA PILHA")
      
    def print(self):
        print("-------")
        celula = self.top
        while celula != None:
          print(celula.item)
          celula = celula.next
        print("-------")

    def seeTop(self):
      if self.tam > 0:
        celula = self.top
        item = celula.item
        return item
      else:
        return None
    
    def isEmpty(self):
      return True if self.tam == 0 else False

def main():
  pilha_p = Stack()
  pilha_aux = Stack()

  lista_itens = [15,12,8,10,14]

  for item in lista_itens:
      if pilha_p.isEmpty() or item >= pilha_p.seeTop():
          pilha_p.insert(item)
      else:
          while not pilha_p.isEmpty() and pilha_p.seeTop() > item:
              pilha_aux.insert(pilha_p.remove())
          pilha_p.insert(item)
          while not pilha_aux.isEmpty():
              pilha_p.insert(pilha_aux.remove())
      pilha_p.print()

main()
````

**Uma palavra ou frase é um palíndromo quando
pode ser lida de trás pra frente mantendo o
mesmo sentido. Exemplos de palavras são “ovo”,
“radar”, “reviver”. Exemplos de frases são “A base
do teto desaba”, “A dama admirou o rim da
amada”.
Faça um programa que recebe uma palavra ou
frase e responda se ela é um palíndromo ou não.**

````python
class Stack:

  def __init__(self):
    self.items = []

  def push(self, data):
    self.items.append(data)

  def is_empty(self):
    return self.items == []

  def pop(self):
    return self.items.pop()


def main():
   stack = Stack()
   text = input("Digite um texto: ").lower()

   for character in text:         #puxando cada caractere para dentro da classe
       stack.push(character)

   reversed_text = ''

   while not stack.is_empty():
      reversed_text = reversed_text + stack.pop()      #tirando da lista o ultimo elemento e colocando na string reversed text

   if text == reversed_text:
      print("É um palíndromo")
   else:
      print("Não é um palíndromo")

main()
````
